Formula of linear Regression ---  Y = mx+c

Y= dependent 

x=independent

m=slop

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/Salary_Age_Data.csv')
df.head()
```

```python
x=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
x_test
```

```python
x_train
```

```python
from sklearn.linear_model import LinearRegression
rig=LinearRegression()
rig.fit(x_train , y_train)
```

```python
y_pred = rig.predict(x_test)
y_pred
```

```python
y_pred[4]
```

```python
y_test[4]
```

```python
rig.score(x_test,y_test)
```

```python
plt.figure(figsize=(10,10))
plt.scatter(x_train,y_train,color='red')
plt.plot(x_train,rig.predict(x_train))
plt.xlabel('Experience')
plt.ylabel('Salary')
plt.title('Salary vs Experience')
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.scatter(x_test,y_test,color='red')
plt.plot(x_train,rig.predict(x_train),color='green')
plt.xlabel('Experience')
plt.ylabel('Salary')
plt.title('Salary vs Experience')
plt.show()
```
